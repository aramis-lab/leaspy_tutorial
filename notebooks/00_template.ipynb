# Tutorial template & `tutolib` demo

This short notebook shows the **two interactive helpers** every tutorial notebook
uses, so you can copy the pattern when authoring your own part:

1. `tp.runquestion(<id>)` — a clickable multiple-choice question with feedback.
2. `tp.solution(<id>)` — reveals a fill-in answer and can **run it for you**.

Run the cells top to bottom. **On Colab the first cell installs everything and
restarts the runtime once** — when it reconnects, run it again (or *Runtime ▸ Run
all*).

```python
# --- Setup -------------------------------------------------------------------
# On Google Colab: installs the (numpy<2) stack leaspy needs, clones the repo,
# then restarts the runtime ONCE so the fresh numpy/pandas load cleanly. When it
# reconnects, just run this cell again (or Runtime > Run all). Running locally
# from inside the repo, it only imports.
import importlib.util, os, pathlib, subprocess, sys

IN_COLAB = importlib.util.find_spec("google.colab") is not None

if IN_COLAB and importlib.util.find_spec("leaspy") is None:
    pip = [sys.executable, "-m", "pip", "install", "-q"]
    subprocess.run(pip + ["leaspy==2.1.*"], check=True)
    # Colab ships numpy 2, but leaspy needs numpy<2. Reinstall the numpy-compiled
    # packages so their binaries match numpy<2 (fixes the dtype-size ABI crash).
    subprocess.run(pip + ["--force-reinstall",
                          "numpy==1.26.4", "pandas==2.2.2",
                          "scipy", "scikit-learn"], check=True)
    if not pathlib.Path("leaspy_tutorial").exists():
        subprocess.run(["git", "clone", "-q",
                        "https://github.com/aramis-lab/leaspy_tutorial"], check=True)
    print("Installed. Restarting the runtime - run this cell again when it reconnects.")
    os.kill(os.getpid(), 9)               # one-time restart; a no-op afterwards

if IN_COLAB:
    sys.path.insert(0, "leaspy_tutorial")
else:
    for _root in (pathlib.Path.cwd(), *pathlib.Path.cwd().parents):
        if (_root / "tutolib").is_dir():
            sys.path.insert(0, str(_root)); break

import tutolib as tp
print("tutolib", tp.__version__, "loaded")
```
