# Tutorial template & `tutolib` demo

>this is meant to be run ONLY in google colab, link bellow

This short notebook shows the **two interactive helpers** every tutorial notebook
uses, so you can copy the pattern when authoring your own part:

1. `tp.runquestion(<id>)` — a clickable multiple-choice question with feedback.
2. `tp.solution(<id>)` — reveals a fill-in answer and can **run it for you**.

Run the cells top to bottom (*Runtime ▸ Run all*). The first cell sets things up
on Colab in a few seconds — no runtime restart needed.

Collab reads directly in the github's repo, so you can just go here to test this notebook in collab

https://colab.research.google.com/github/aramis-lab/leaspy_tutorial/blob/main/notebooks/00_template.ipynb


```python
# --- Setup -------------------------------------------------------------------
# On Colab: install leaspy WITHOUT touching Colab's preinstalled numpy/torch
# (--no-deps), clone this repo, and load the helper library. No downgrade, no
# restart. Running locally, it only imports.
#
# DEVS: to test your OWN branch, set BRANCH below to it AND open this notebook
# from that branch (File > Open notebook > GitHub > branch dropdown). Leave
# BRANCH = "main" on the main branch. To re-pull after pushing new commits,
# delete the leaspy_tutorial folder (or factory-reset the runtime).
BRANCH = "main"

import importlib.util, pathlib, subprocess, sys

IN_COLAB = importlib.util.find_spec("google.colab") is not None
pip = [sys.executable, "-m", "pip", "install", "-q"]

if IN_COLAB:
    if importlib.util.find_spec("leaspy") is None:
        subprocess.run(pip + ["--no-deps", "leaspy==2.0.2"], check=True)
        subprocess.run(pip + ["lifelines"], check=True)   # dep Colab may lack
    if not pathlib.Path("leaspy_tutorial").exists():
        subprocess.run(["git", "clone", "-q", "--branch", BRANCH,
                        "https://github.com/aramis-lab/leaspy_tutorial"], check=True)
    sys.path.insert(0, "leaspy_tutorial")
else:
    for _root in (pathlib.Path.cwd(), *pathlib.Path.cwd().parents):
        if (_root / "tutolib").is_dir():
            sys.path.insert(0, str(_root)); break

import tutolib as tp
print("tutolib", tp.__version__, "loaded")
```

## 2. Fill-in exercises
